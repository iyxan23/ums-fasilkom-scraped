```python
import json

material_content_file = "scraped-content.json"
material_content = []
materials_file = "scraped-materials.csv"

with open(material_content_file, "r") as f:
    material_content = json.load(f)
```

```python
import pandas as pd

materials = pd.read_csv(materials_file, index_col="Code")
materials.head()
```

```python
import json
import pandas as pd

def extract_title_and_clean_content(content):
    from bs4 import BeautifulSoup

    soup = BeautifulSoup(content, "html.parser")

    text_content = []
    for element in soup.find_all(["h2", "h3", "p", "li", "div"]):
        text = element.get_text(separator="\n").strip()
        if text:
            text_content.append(text)

    return "\n".join(text_content)

material_content_dict = {}

for item in material_content:
    code = item["code"]
    content = extract_title_and_clean_content(item["content"])

    material_content_dict[code] = content

materials["Content"] = materials.index.map(material_content_dict)

materials.head()
```

```python
materials.to_csv("ums-fasilkom.csv")
```
